```python
import torch, torchvision
```

```python
# Import a resnet18 pretrained model
model = torchvision.models.resnet18(pretrained=True)
```

```python
# Define the data (1 image with 3 color channels that is 64x64) and the labels
data = torch.rand(1, 3, 64, 64)
labels = torch.rand(1, 1000)
```

```python
# Predictions from the model over the data.
prediction = model(data)
# Loss as a sum of difference betwee the actual and the expected.
loss = (prediction - labels).sum()
# Backprop on the parameters of the model (Stored in each parameters .grad)
loss.backward()
loss
```

```python
for i in model.parameters():
    print(i.grad)
```

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
optim.step()
```

```python
# Differentiation in Autograd

# Assume a, b are parameters in a NN
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)

# Assume Q is the error at the end of a NN
Q = 3*a**3 - b**2
```

```python
external_grad = torch.tensor([1., 1.])
Q.backward(external_grad)
```

```python
print(a.grad == 9*a**2)
print(b.grad == -2*b)
```
